```python
#coordinates of H2, H2O, and benzene
coor_H2 = {
    "H1" : [0, 0, 0],    
    "H2" : [0, 0, 0.7414]
}
print("coordinates of H2:", coor_H2)

coor_H2O = {
    "H1" : [0, 0, 0.1173], 
    "H2" : [0, 0.7572, -0.4692],
    "O" : [0, -0.7572, -0.4692]
}
print("coordinates of H2O:", coor_H2O)

coor_benzene = {
    "C1" : [0.0000,	1.3970,	0.0000],
    "C2" : [1.2098, 0.6985, 0.0000],
    "C3" : [1.2098,	-0.6985, 0.0000],
    "C4" : [0.0000, -1.3970, 0.0000],
    "C5" : [-1.2098, -0.6985, 0.0000],
    "C6" : [-1.2098, 0.6985, 0.0000],
    "H7" : [0.0000,	2.4810,	0.0000],
    "H8" : [2.1486,	1.2405,	0.0000],
    "H9" : [2.1486,	-1.2405, 0.0000],
    "H10" : [0.0000, -2.4810, 0.0000],
    "H11" : [-2.1486, -1.2405, 0.0000],
    "H12" : [-2.1486, 1.2405, 0.0000]
}
print("coordinates of benzene:", coor_benzene)

# function to calculate the bond length
def compute_bond_length(coor1, coord2):
        
#calculate bond length in armstrong  
    bond_length = np.sqrt((coor1[0] - coord2[0]) ** 2 + (coor1[1] - coord2[1]) ** 2 + (coor1[2] - coord2[2]) ** 2)
    return bond_length 
    
bond_length = compute_bond_length(coor_benzene['H7'], coor_benzene['H8'])  

print("The bond length is", bond_length)
 
#check if the bond length is too long    
if bond_length <= 2:
    print("Its a covalent bond")
    
else:
    print("The bond distance is too long!")

# Function to calculate the bond angle
def compute_bond_angle(coord1, coord2, coord3):
    
    # Vector from atom1 to atom2 and atom2 to atom3
    vector_atom_1_2 = [coord2[i] - coord1[i] for i in range(3)]
    vector_atom_2_3 = [coord3[i] - coord2[i] for i in range(3)]
    
    # Magnitudes of the vectors
    mag_1 = np.sqrt(sum([(coord2[i] - coord1[i]) ** 2 for i in range(3)]))
    mag_2 = np.sqrt(sum([(coord3[i] - coord2[i]) ** 2 for i in range(3)]))
       
    # Dot product of the two vectors
    dot_product= sum([vector_atom_1_2[i] * vector_atom_2_3[i] for i in range(3)])
    
    # Calculate the bond angle
    bond_angle = np.arccos(dot_product / (mag_1 * mag_2))
    return bond_angle   
    
bond_angle = compute_bond_angle(coor_H2O['H1'], coor_H2O['H2'], coor_H2O['O']) 

# convert the bond angles to degrees
bond_angle_deg = np.degrees(bond_angle)
print(bond_angle_deg) 

# check the type of bond angle
if bond_angle_deg < 90:
    print("Acute angle")
    
elif bond_angle_deg == 90:
    print("Right angle")
    
else:
    print("Obtuse angle")   

def calculate_all_bond_lengths(molecule):
    bond_lengths = []
    atoms = list(molecule.keys())
    
    for i in range(len(atoms)):
        for j in range(i + 1, len(atoms)):  
            atom1 = atoms[i]
            atom2 = atoms[j]
            bond_length = compute_bond_length(molecule[atom1], molecule[atom2])
            bond_lengths.append((atom1, atom2, bond_length)) 
    return bond_lengths

# Calculate all bond lengths
all_bond_lengths = calculate_all_bond_lengths(coor_benzene)
print('all_bond_lengths', all_bond_lengths)

def calculate_all_bond_angles(molecule):
    bond_angles = []
    atoms = list(molecule.keys())
    for i in range(len(atoms)):
        for j in range(i + 1, len(atoms)):  
            for k in range(j + 1, len(atoms)):
                atom1 = atoms[i]
                atom2 = atoms[j]
                atom3 = atoms[k]
                bond_angle = compute_bond_angle(molecule[atom1], molecule[atom2], molecule[atom3])
                bond_angles.append((atom1, atom2, atom3, bond_angle)) 
    return bond_angles

# Calculate all bond lengths
all_bond_angles = calculate_all_bond_angles(coor_benzene)  
print('all_bond_angles', all_bond_angles)
```
